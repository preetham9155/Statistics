# 2 Proportion t-Test - From statsmodels.api -stats.ttest_ind

```python
n1=247
p1=.37
n2=308
p2=.39
```

```python
sample1 = np.random.binomial(1,p1,n1) 
# it will n1(247) data and make p1(37%) in them 
# 1 is given beacuse it will take that number create 37%
```

```python
sample1
```

```python
sample2=np.random.binomial(1,p2,n2)
```

```python
sample2
```

```python
len(sample1)
```

```python
len(sample2)
```

```python
import statsmodels.api as sm
p=sm.stats.ttest_ind(sample1,sample2)[1]
print(p)
print(round(p,3))
```

# Chi - Square Test

```python
from scipy.stats import chi2_contingency
from matplotlib import pyplot as plt
import pandas as pd
```

```python
data=[[14,4],[0,10]]
data
```

```python
df=pd.DataFrame(data,
                index=['Athlete','Non-Athlete'],
                columns=['Non-smoker','smoker'])
```

```python
df
```

```python
df.plot(kind='bar',stacked=False)
```

```python
# test statistic , p-value,df,expected freuency
chi2,p,dof,exp=chi2_contingency(data)
```

```python
chi2_contingency(data)
```

```python
chi2
```

```python
p
```

```python
dof # degree of freedom
```

```python
exp # expected frequency
```

```python
# compare p with alpha
# p=0.0003857 , alpha = 0.05
# p<alpha -> Rejected Null Hypothesis
# Accepted Alternative hypothesis
```

```python
a=0.05
if p<a:
    print("Reject null hypothesis ")
else:
    print("Accept alternative hypothesis")
```
